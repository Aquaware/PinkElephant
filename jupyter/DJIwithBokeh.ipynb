```python
from math import pi
import pandas as pd
from bokeh.io import output_notebook, export_png 
#from bokeh.layout import columns
from bokeh.plotting import figure, show, output_file
output_notebook()
import Market


if __name__ == '__main__':
    brand = 'dji'
    dji = Market.Market(brand, 5, Market.UNIT_MINUTE, False)
    dji.importFromSQLite('./data/click_sec/' + brand + '_5min.db', brand)
    print('imported data size ...', len(dji.df))
    tbegin = Market.DTime(2019, 8, 5, 16, 30)
    market = dji.timeRangeFilter(tbegin, tbegin + Market.DeltaDay(1))
    TOOLS = "pan,wheel_zoom,box_zoom,reset,save"

    p = figure(x_axis_type="datetime", tools=TOOLS, plot_width=1000, title = "MSFT Candlestick")
    p.xaxis.major_label_orientation = pi/4
    p.grid.grid_line_alpha=0.3

    df = market.df
    inc = df.close > df.open
    dec = df.open > df.close
    w = 5 * 60 * 1000 # 5min = 60sec * 1000msec
    p.segment(df.time, df.high, df.time, df.low, color="black")
    p.vbar(df.time[inc], w, df.open[inc], df.close[inc], fill_color="#D5E1DD", line_color="black")
    p.vbar(df.time[dec], w, df.open[dec], df.close[dec], fill_color="#F2583E", line_color="black")

    #output_file("candlestick.html", title="candlestick.py example")

    show(p)  # open a browser
```
